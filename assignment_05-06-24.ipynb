```python
skills = data["Keywords"].values
skills.shape
```

```python
# This code snippet is creating a dictionary called `skills_dict` to store skills and their frequencies. It iterates over a list of skills, splits each skill by comma, strips any leading or trailing spaces, and converts it to lowercase. It then checks if the skill is already in the `skills_dict`. If it is, it increments the count of that skill by 1. If the skill is not in the dictionary, it adds the skill to the dictionary with a count of 1. This way, the dictionary keeps track of how many times each skill appears in the list of skills.
skills_dict = {}
for skill in skills:
    for s in skill.split(","):
        s = s.strip().lower()
        if s in skills_dict:
            skills_dict[s] += 1
        else:
            skills_dict[s] = 1
```

```python
skills_dict = dict(sorted(skills_dict.items(), key=lambda x: x[1], reverse=True)) 
skills_dict
```

```python
# make a bar graph showing the top five skills sought after in the market
skills_df = pd.DataFrame(skills_dict.items(), columns=["Skill", "Count"]).head(5)
plt.figure(figsize=(10, 10))
plt.bar(skills_df["Skill"], skills_df["Count"], color= "orange")
plt.title("Top five skills sought after in the market")
plt.show()
```

```python
# This code snippet is iterating over each row in the `data` DataFrame and checking if any of the skills from the `skills_dict` dictionary are present in the "Keywords" column of that row. If a skill is found, it is added to the "top_skills" column of that row. The loop stops when 6 skills have been added to the "top_skills" column for that row.
data["top_skills"] = ""

for i in range(data.shape[0]):
    for skill in skills_dict.keys():
        if skill.strip() in data.iloc[i,]["Keywords"].strip().lower():
            data.iloc[i, -1] += skill + ", "
            if len(data.iloc[i, -1].split(",")) == 6:
                break
    if data.iloc[i, -1] == "":
        data.iloc[i, -1] = "No Skill"
    else:
        data.iloc[i, -1] = data.iloc[i, -1][:-2]

        
data["top_skills"]
```
